## In order for us to be able to analyze the correlation better in this data, we visualize the data of correlation values.
* Correlation value close to 1 means positive correlation
* Correlation value close to -1 means negative correlation
* Correlation value close to 0 means the data has no correlation. For the sake of analysis, we will define correlation value under 0.2, -0.2 as no correlation.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Compute the correlation matrix
corr = recent_grads.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(20,14))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
ax = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot = True)
```

## Comparing the number of men with the number of women in each category of majors.

```python
category_uniq = recent_grads["Major_category"].unique()
cat_women = {}
cat_men = {}
for cat in category_uniq:
    boolcat = recent_grads['Major_category'] == cat
    cat_w = recent_grads[boolcat]['Women']
    cat_m = recent_grads[boolcat]['Men']
    cat_women[cat] = cat_w.sum()
    cat_men[cat] = cat_m.sum()

df_cat= pd.DataFrame({'Women':cat_women, 'Men': cat_men})
df_cat = df_cat.sort_values(by='Women')

df_cat.plot.bar()
```

## Most popular major for Women are: Business, Education, and Liberal Arts
## Most popular major for Men are: Business, Engineering, Liberal Arts

```python
def maj_cat_mean(s):
    category_uniq = recent_grads["Major_category"].unique()
    cat_dict = {}
    for cat in category_uniq:
        boolcat = recent_grads['Major_category'] == cat
        cat_1 = recent_grads[boolcat][s]
        cat_dict[cat] = cat_1.mean()
    return cat_dict
def maj_cat_max(s):
    category_uniq = recent_grads["Major_category"].unique()
    cat_dict = {}
    for cat in category_uniq:
        boolcat = recent_grads['Major_category'] == cat
        cat_1 = recent_grads[boolcat][s]
        cat_dict[cat] = cat_1.max()
    return cat_dict
```

```python
unemp_dict = maj_cat_mean('Unemployment_rate')
df_cat_unem= pd.DataFrame({'Unemployment rate':unemp_dict})
df_cat_unem = df_cat_unem.sort_values(by='Unemployment rate')
df_cat_unem.plot.bar(title = 'Average Unemployment Rating per Major Category', legend = False)
```

```python
unemp_dict = maj_cat_max('Unemployment_rate')
df_cat_unem= pd.DataFrame({'Unemployment rate':unemp_dict})
df_cat_unem = df_cat_unem.sort_values(by='Unemployment rate')
df_cat_unem.plot.bar(title = 'Max Unemployment Rating per Major Category', legend = False)
```

```python
eng_cat = recent_grads[recent_grads['Major_category'] == 'Engineering'].sort_values(by='Unemployment_rate')
eng_cat_major = eng_cat[eng_cat['Unemployment_rate']==eng_cat['Unemployment_rate'].max()]
eng_cat.plot.bar(x='Major', y='Unemployment_rate', legend = False)
```

```python
average = str(eng_cat['Unemployment_rate'].mean())
print('Engineering Average unemployment rate: ' + average )
print('Engineering major with highest unemployment rate: '+ str(eng_cat_major['Major']))
print('Ranking of highest unemployment rate: '+ str(eng_cat_major['Rank']))
```

```python
tot_dict = maj_cat_mean('Total')
df_cat_tot= pd.DataFrame({'Total':tot_dict})
df_cat_tot = df_cat_tot.sort_values(by='Total')
df_cat_tot.plot.bar(title = 'Average Total Students per Major Category', legend = False)
```
